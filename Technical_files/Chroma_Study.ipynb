```python
import chromadb

# Create a client
client = chromadb.Client()

# Create a collection (like a table in traditional databases)
collection = client.create_collection(name="company_docs")


# Add documents with their IDs and optional metadata
collection.add(
    documents=[
        "Our return policy allows returns within 30 days",
        "We offer free shipping on orders over $50",
        "Customer support is available 24/7 via chat"
    ],
    ids=["doc1", "doc2", "doc3"],
    metadatas=[
        {"category": "returns"},
        {"category": "shipping"},
        {"category": "support"}
    ]
)

# Search for relevant documents
results = collection.query(
    query_texts=["How do I return an item?"],
    n_results=3
)

print(results['documents'])
print(results['metadatas'])
print()
print(results['ids'])
print()
print(results['distances'])
print()
print(results['embeddings'])
print()
print(results['documents'][0][0])

# Output: [["Our return policy allows returns within 30 days", ...]]
```
