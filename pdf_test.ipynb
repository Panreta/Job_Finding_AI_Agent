# Reading the CV of mine and transverse it into txt.

```python

import os
import pypdf  


def read_resume():
    """Read resume from data folder (supports PDF and TXT)"""
  
    
    # Try PDF first
    pdf_path = "D:\\Github\\Job_Finding_AI_Agent\\data\\CV.pdf"
    if os.path.exists(pdf_path):
        try:
            with open(pdf_path, 'rb') as f:
                reader = pypdf.PdfReader(f)
                text = ""
                for page in reader.pages:
                    text += page.extract_text()
                return text
        except Exception as e:
            return f"Error reading PDF: {str(e)}"
    
    # Try TXT as fallback
    txt_path = "data/resume.txt"
    if os.path.exists(txt_path):
        with open(txt_path, 'r', encoding='utf-8') as f:
            return f.read()
    
    return "Resume not found. Please add data/resume.pdf or data/resume.txt"

print(read_resume())
```
